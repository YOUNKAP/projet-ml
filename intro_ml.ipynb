## 1 - importation des bibliothèques 

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

<hr/> 

## 2 - Création séries de 10 prédictions simulées

```python
np.random.seed(42)  # pour reproductibilité

# Vraies valeurs
y_true = pd.Series(np.random.normal(loc=100, scale=10, size=10))

# Prédictions simulées
y_pred = y_true + np.random.normal(loc=0, scale=5, size=10)

print("Vraies valeurs :")
print(y_true)

print("\nPrédictions :")
print(y_pred)
```

<hr/> 

## 3 - Visualisation

### 3.1 - Histogramme

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calcul des erreurs
erreurs = y_pred - y_true

plt.figure()
sns.histplot(erreurs, bins=8, kde=True)
plt.title("Distribution des erreurs")
plt.xlabel("Erreur")
plt.ylabel("Fréquence")
plt.show()
```

### 3.2 - Nuage

```python
plt.figure()
plt.scatter(y_true, y_pred)
plt.plot([y_true.min(), y_true.max()],
         [y_true.min(), y_true.max()])
plt.xlabel("Valeurs réelles")
plt.ylabel("Prédictions")
plt.title("Réel vs Prédit")
plt.show()
```

<hr/> 

## 4 - différence entre surapprentissage  et sous-apprentissage

<div class="alert alert-info" style="font-size:100%">
<b>Surapprentissage (Overfitting) </b>: <br><br>
    
Le modèle est trop **complexe** et **mémorise** les données d’entraînement avec le bruit 
* Mauvaises performances sur les **données d’entraînement**
* Mauvaises performances sur les **données de test**
* **Variance** élevée

</div>

<div class="alert alert-success" style="font-size:100%">
<b>Sous-apprentissage (Underfitting)  </b>: <br><br>

Le modèle est **trop simple** pour capturer la structure des données.

* Excellente performance sur les **données d’entraînement**
* Mauvaises performances sur les **données de test**
* **Bias** élevée

</div>
